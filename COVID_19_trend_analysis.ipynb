```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
df = pd.read_csv(file_name)

```

```python
print(df.head())


```

```python
# Summary statistics of the dataset
summary_stats = df.describe()

# Plot the summary statistics
summary_stats.T.plot(kind='bar', figsize=(10, 6))
plt.title('Summary Statistics of COVID-19 Dataset')
plt.ylabel('Value')
plt.show()
```

```python
# Group by 'WHO Region' and calculate the total confirmed, deaths, and recovered cases
grouped_data = df.groupby('WHO Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
}).reset_index()
```

```python
# Select the top 20 major regions based on the total confirmed cases
top_regions = grouped_data.nlargest(20, 'Confirmed')
```

```python
# Plot the total confirmed, deaths, and recovered cases for the top 20 regions
top_regions.plot(kind='bar', x='WHO Region', y=['Confirmed', 'Deaths', 'Recovered'])
plt.title('Top 20 Major Regions - COVID-19 Statistics')
plt.ylabel('Total Cases')
plt.show()
```
